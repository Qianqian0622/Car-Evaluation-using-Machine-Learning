```python
# MODIFY THIS CELL

def draw_bootstrap_sample(rng, X, y):
    sample_indices = np.arange(0,1243)
    bootstrap_indices = rng.choice(sample_indices,1243,replace=True)
    return X[bootstrap_indices], y[bootstrap_indices]

```

```python
rng = np.random.RandomState(123)
X_boot, y_boot = draw_bootstrap_sample(rng, X_train, y_train)

print('Number of training inputs from bootstrap round:', X_boot.shape[0])
print('Number of training labels from bootstrap round:', y_boot.shape[0])
print('Labels:\n', y_boot)
```

```python
from sklearn.neighbors import KNeighborsClassifier


class BaggingClassifier(object):
    
    def __init__(self, num_knn=10, random_state=123):
        self.num_knn = num_knn
        self.rng = np.random.RandomState(random_state)
        
        
    def fit(self, X, y):
        self.knn_=[]
        for j in range(16):
            self.knn_.append([KNeighborsClassifier(n_neighbors=j+1) for i in range(self.num_knn)])
        for i in range(16):
            for j in range(self.num_knn):
                X_boot, y_boot = draw_bootstrap_sample(self.rng, X, y)# YOUR CODE to draw a bootstrap sample
                self.knn_[i][j].fit(X_boot, y_boot)# YOUR CODE to
            # fit the trees in self.trees_ on the bootstrap samples
        
    def predict(self, X):
        ary = np.zeros((X.shape[0],16,10), dtype=np.int)
        for i in range(16):
            for j in range (self.num_knn):
                ary[:, i,j] = self.knn_[i][j].predict(X)

        maj = np.apply_along_axis(lambda x:
                                  np.argmax(np.bincount(x)),
                                            axis=2,
                                            arr=ary)
        return maj
```

```python
from sklearn.tree import DecisionTreeClassifier


class BaggingClassifier(object):
    
    def __init__(self, num_trees=10, random_state=123):
        self.num_trees = num_trees
        self.rng = np.random.RandomState(random_state)
        
        
    def fit(self, X, y):
        self.trees_ = [DecisionTreeClassifier(random_state=self.rng) for i in range(self.num_trees)]
        for i in range(self.num_trees):
            X_boot, y_boot = draw_bootstrap_sample(self.rng, X_train, y_train)
            self.trees_[i].fit(X_boot,y_boot)
        
    def predict(self, X):
        ary = np.zeros((X.shape[0], len(self.trees_)), dtype=np.int)
        for i in range(len(self.trees_)):
            ary[:, i] = self.trees_[i].predict(X)

        maj = np.apply_along_axis(lambda x:
                                  np.argmax(np.bincount(x)),
                                            axis=1,
                                            arr=ary)
        return maj
```

```python
model = BaggingClassifier()
model.fit(X_train, y_train)

predictions = model.predict(X_valid)

print('Individual Tree Accuracies:')
for tree in model.trees_:
    predictions = tree.predict(X_valid) 
    print('%.1f%%' % ((predictions == y_valid).sum() / X_valid.shape[0] * 100))

print('\nBagging Valid Accuracy: %.1f%%' % ((predictions == y_valid).sum() / X_valid.shape[0] * 100))
```

# 10- fold CV for decision tree

```python
from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold(n_splits=10, random_state=123, shuffle=True)

kfold_acc = 0.
kfold_values=[]
for train_idx, valid_idx in cv.split(X_train, y_train):
    clf = DecisionTreeClassifier(random_state=123, max_depth=9).fit(X_train[train_idx], y_train[train_idx])
    y_pred = clf.predict(X_train[valid_idx])
    acc = np.mean(y_pred == y_train[valid_idx])*100
    kfold_acc += acc
    kfold_values.append(acc)
kfold_acc /= 10
    
clf = DecisionTreeClassifier(random_state=123, max_depth=9).fit(X_train, y_train)
y_pred = clf.predict(X_test)
test_acc = np.mean(y_pred == y_test)*100
###数据量越小，越容易overfit    
print('Kfold Accuracy: %.2f%%' % kfold_acc,np.std(kfold_values))
print('Test Accuracy: %.2f%%' % test_acc)

```
